```python
import math 
import numpy as np
t_max = 10 # maximum time
h = 0.1 # time step
g = -9.81
n_steps = int(t_max/h)
t = np.zeros(n_steps)
v = np.zeros(n_steps)
 
# set initial conditions
t[0] = 0
v[0] = 0
 
for i in range(1,n_steps):
    t[i] = t[i-1] + h
    v[i] = v[i-1] + h * g
print t
print v
    
```
