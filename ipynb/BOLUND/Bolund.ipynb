```python
from zutil.post import for_each

def plot_velocity_profile(ax,file_root):
    
    farfield = PVDReader( FileName=file_root+'_farfield.pvd' )
    farfield.UpdatePipeline()
    
    merged = MergeBlocks(Input=farfield)
    merged.UpdatePipeline()

    point_data = CellDatatoPointData(Input=merged)
    point_data.PassCellData = 0
    point_data.UpdatePipeline()

    clip = Clip( ClipType="Plane", Input=point_data )
    clip.ClipType.Normal = [1.0, 0.0, 0.0]
    clip.ClipType.Origin = [0.0, 0.0, 0.0]    
    clip.UpdatePipeline()


    wall_slice = Slice(Input=clip, SliceType="Plane" )
    wall_slice.SliceType.Normal = [-1.0,1.0,0.0]
    wall_slice.SliceType.Origin = [-90, -90.0, 0.0]
    wall_slice.UpdatePipeline()
    
    sorted_line = PlotOnSortedLines(Input=wall_slice)
    sorted_line.UpdatePipeline()
    
    slice_client = servermanager.Fetch(sorted_line)
    for_each(slice_client,func=plot_array,axis=ax)         

def plot_array(data_array,pts_array,**kwargs):
    ax = kwargs['axis']
    data = []
    count = 0
    u_vel = data_array.GetPointData()['V'][:,0]
    v_vel = data_array.GetPointData()['V'][:,1]
    w_vel = data_array.GetPointData()['V'][:,2]
    for p in pts_array.GetPoints()[:,0]:
        u = float(u_vel[count])
        v = float(v_vel[count])
        w = float(w_vel[count])        
        vel_mag = math.sqrt(u*u + v*v + w*w)
        data.append(vel_mag)
        count+=1 
    ax.plot(data, pts_array.GetPoints()[:,2], color='g',linestyle='-',marker='None',label='Actual')
    
def plot_defined_profile(ax):
    surface_layer_height = 1000.0 # varies from 10m to 200m
    roughness = 0.0003 # Open sea
    friction_velocity = 0.4 # Typically 0.2 m/s during the day
    # ABL height
    earth_rotation_speed=465.0#m/s
    earth_rotation_speed=0.0000727221 #rad/s
    latitude = math.radians(20.0)
    z_h = 0.25*(friction_velocity/(2*math.pi*earth_rotation_speed*math.sin(latitude)))
    #print z_h
    z0 = -0.75
    KarmanConst = 0.41
    pos = []
    data = []
    for z in range(0,1000):
        vel = friction_velocity/KarmanConst*math.log((float(z)-z0+roughness)/roughness)
        if z >= surface_layer_height:
            vel = 0.03*math.sqrt(1.4*287*(273.15+15.0))
        data.append(vel)
        pos.append(float(z)-z0)
    ax.plot(data, pos, color='r',linestyle='-',marker='None',label='Specified')
```

```python
from zutil.post import ProgressBar
pb = ProgressBar()

# Read measured points
output_points_txt = CSVReader( FileName=['output_points.txt'] )
output_points_txt.MergeConsecutiveDelimiters = 1
output_points_txt.FieldDelimiterCharacters = ', '
output_points_txt.HaveHeaders = 0
output_points_txt.UpdatePipeline()

# Convert table to points
TableToPoints1 = TableToPoints( Input = output_points_txt)
TableToPoints1.XColumn = 'Field 0'
TableToPoints1.YColumn = 'Field 1'
TableToPoints1.ZColumn = 'Field 2'
TableToPoints1.UpdatePipeline()

pb += 10

from zutil.post import get_case_root
vtu = PVDReader( FileName=[get_case_root(case_name,num_procs)+'.pvd'] )
vtu.UpdatePipeline()
pb += 10
merged = CleantoGrid(Input=vtu)
merged.UpdatePipeline()
pb += 10
point_data = CellDatatoPointData(Input=merged)
point_data.PassCellData = 0
point_data.PieceInvariant = 1
point_data.UpdatePipeline()
pb += 10
#ResampleWithDataset1 = ResampleWithDataset( Input=point_data, Source=TableToPoints1 )
#ResampleWithDataset1.UpdatePipeline()

from paraview.vtk.dataset_adapter import DataSet
#resample = servermanager.Fetch(ResampleWithDataset1)
#resample_data = DataSet(resample)
pb += 10

# Read in measured points
measured_points_txt = CSVReader( FileName=['Dir_239.dat'] )
measured_points_txt.MergeConsecutiveDelimiters = 1
measured_points_txt.FieldDelimiterCharacters = ', '
measured_points_txt.HaveHeaders = 1
measured_points_txt.UpdatePipeline()

# Convert table to points
TableToPoints2 = TableToPoints(Input=measured_points_txt)
TableToPoints2.XColumn = 'x'
TableToPoints2.YColumn = 'y'
TableToPoints2.ZColumn = 'z'
TableToPoints2.UpdatePipeline()
pb += 10

measured = servermanager.Fetch(TableToPoints2)
measured_data = DataSet(measured)

pb.complete()
```

```python
from zutil.post import ProgressBar
pb = ProgressBar()
velocity_probe=[]
probe=ProbeLocation( Input=point_data, ProbeType="Fixed Radius Point Source" )
for p in range(measured_data.GetNumberOfPoints()):
    pt=measured_data.GetPoint(p)
    #print pt
    probe.ProbeType.Center = list(pt)
    probe.UpdatePipeline()
    probe_client = servermanager.Fetch(probe)
    probe_data = DataSet(probe_client)
    u_pvel = probe_data.GetPointData()['V'].GetTuple(0)[0]
    v_pvel = probe_data.GetPointData()['V'].GetTuple(0)[1]
    w_pvel = probe_data.GetPointData()['V'].GetTuple(0)[2]

    velocity_probe.append([u_pvel,v_pvel,w_pvel])
    
    pb += math.ceil(1.0/float(measured_data.GetNumberOfPoints())*100.0)
pb.complete()
```

```python

probe_vel_mag=[]
probe_uvel=[]
probe_vvel=[]
probe_wvel=[]
measured_vel_mag=[]
measured_uvel=[]
measured_vvel=[]
measured_wvel=[]
measured_id=[]
measured_component_id=[]
for id,v in enumerate(velocity_probe):
    #vel = measured_data2.GetPointData()['Vm'].GetTuple(id)
    u_t = measured_data.PointData[4][id]
    vel_ut = measured_data.PointData[5][id]
    um = measured_data.PointData[6][id]*u_t
    vm = measured_data.PointData[7][id]*u_t
    wm = measured_data.PointData[8][id]*u_t
    #print v[0][0], um, v[0][1], vm , v[0][2], wm
    #print v
    
    measurement_name = measured_data.PointData[0].GetValue(id)

    probe_vel_mag.append(math.sqrt(v[0]*v[0]+v[1]*v[1]+v[2]*v[2]))
    measured_vel_mag.append(vel_ut*u_t)
    measured_id.append(measurement_name)
    
    if measurement_name.endswith('S'):
        probe_uvel.append(v[0])
        probe_vvel.append(v[1])
        probe_wvel.append(v[2])
        measured_uvel.append(um)    
        measured_vvel.append(vm)
        measured_wvel.append(wm)
        measured_component_id.append(measurement_name)


    
def autolabel(rects):
    # attach some text labels
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.05*height, '%d'%int(height),
                ha='center', va='bottom')

ind = np.arange(len(measured_id))                # the x locations for the groups
ind2 = np.arange(len(measured_component_id))                # the x locations for the groups
width = 0.35                      # the width of the bars

from zutil.post import get_case_root
fig = pl.figure(figsize=(30, 30),dpi=100, facecolor='w', edgecolor='k')
fig.suptitle('Bolund Mast Comparison - Wind 239deg', fontsize=20, fontweight='bold')
ax = fig.add_subplot(4,1,1)
ax.grid(True)

rects1 = ax.bar(ind, probe_vel_mag, width, color='r')
rects2 = ax.bar(ind+width, measured_vel_mag, width, color='y')
ax.set_ylabel('Speed m/s')
ax.set_title('Speed comparison')
ax.set_xticks(ind+width)
ax.set_xticklabels( measured_id, rotation=90 )
ax.legend( (rects1[0], rects2[0]), ('zCFD', 'Measured') )
#autolabel(rects1)
#autolabel(rects2)


ax = fig.add_subplot(4,1,2)
ax.grid(True)
rects1 = ax.bar(ind2, probe_uvel, width, color='r')
rects2 = ax.bar(ind2+width, measured_uvel, width, color='y')
ax.set_ylabel('u m/s')
ax.set_title('U velocity component comparison')
ax.set_xticks(ind2+width)
ax.set_xticklabels( measured_component_id, rotation=90 )
ax.legend( (rects1[0], rects2[0]), ('zCFD', 'Measured') )

ax = fig.add_subplot(4,1,3)
ax.grid(True)
rects1 = ax.bar(ind2, probe_vvel, width, color='r')
rects2 = ax.bar(ind2+width, measured_vvel, width, color='y')
ax.set_ylabel('v m/s')
ax.set_title('V velocity component comparison')
ax.set_xticks(ind2+width)
ax.set_xticklabels( measured_component_id, rotation=90)
ax.legend( (rects1[0], rects2[0]), ('zCFD', 'Measured') )

ax = fig.add_subplot(4,1,4)
ax.grid(True)
rects1 = ax.bar(ind2, probe_wvel, width, color='r')
rects2 = ax.bar(ind2+width, measured_wvel, width, color='y')
ax.set_ylabel('w m/s')
ax.set_title('W velocity component comparison')
ax.set_xticks(ind2+width)
ax.set_xticklabels( measured_component_id, rotation=90 )
ax.legend( (rects1[0], rects2[0]), ('zCFD', 'Measured') )


#ax.legend(loc='upper right', shadow=True)
show()
```
