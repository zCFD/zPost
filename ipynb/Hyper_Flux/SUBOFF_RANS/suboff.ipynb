```python
from zutil.post import cp_profile_wall_from_file

def plot_cp_profile(ax,file_root,symmetry):
    
    force_data = cp_profile_wall_from_file(file_root,
                              [0.0,0.0,1.0],
                              [0, 0, 0],
                              func=plot_array,
                              axis=ax,
                              symmetry=symmetry,
                              alpha=alpha)
    
def plot_array(data_array,pts_array,**kwargs):
    ax = kwargs['axis']
    symmetry = kwargs['symmetry']
    cp_array = data_array.GetPointData()['cp']
    chord_array = data_array.GetPointData()['chord']
    ax.plot(chord_array, cp_array , 'b.',color='k',label='zCFD SST',markersize=10)
```
